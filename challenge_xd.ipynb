```python
import pandas as pd
import requests
import json
```

### 1.- Conexión al enlace y creación del DataFrame

```python
url = 'https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow'
r = requests.get(url).json()
```

```python
r_items = r['items']
df_items = pd.DataFrame(r_items)
list_owners = list(df_items['owner'])
df_owners = pd.DataFrame(list_owners)
df = pd.concat([df_items,df_owners], axis=1, join='inner')
```

### 2.- Obtener el número de respuestas contestadas y no contestadas

```python
contestado = len(df[df['is_answered'] == True])
no_contestado = len(df[df['is_answered'] == False])
print( "Contestado = " + str(contestado) + " No contestado= " + str(no_contestado))
```

### 3.- Obtener la respuesta con menor número de vistas

```python
min_views = df['view_count'].min()
min_views_answer = df[df['view_count'] == min_views]
min_views_answer
```

### 4.- Obtener la respuesta más vieja y más actual

```python
oldest_answer = df[df['creation_date'] == df['creation_date'].min()]
newest_asnwer = df[df['creation_date'] == df['creation_date'].max()]
```
